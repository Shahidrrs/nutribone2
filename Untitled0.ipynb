```python
import os
import glob
import numpy as np
import kagglehub
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import Sequence

# 1. Fetch data path safely
path = kagglehub.dataset_download("shashwatwork/knee-osteoarthritis-dataset-with-severity")
print("Downloaded raw directory structure:", os.listdir(path))

# DYNAMIC PATH FINDER: Hunt down where the actual splits are stored
all_train_paths = glob.glob(os.path.join(path, "**", "*train*"), recursive=True)
all_val_paths = glob.glob(os.path.join(path, "**", "*val*"), recursive=True)

if all_train_paths:
    train_dir = all_train_paths[0]
    # Match validation directory structurally
    val_dir = all_val_paths[0] if all_val_paths else train_dir.replace("train", "val")
else:
    # Manual fallback check to list contents and handle arbitrary structure anomalies
    raise FileNotFoundError(f"Could not automatically locate subfolder paths under {path}. Current layout: {os.listdir(path)}")

print("\n--- RESOLVED REAL PATHS ---")
print("Target Train Directory mapped to:", train_dir)
print("Target Val Directory mapped to:", val_dir)

# Regular, safe geometric data generators
train_datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    brightness_range=[0.85, 1.15],
    horizontal_flip=True,
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator()

# Load all 5 classes from disk cleanly based on verified paths
raw_train_gen = train_datagen.flow_from_directory(
    train_dir, target_size=(224, 224), batch_size=32, class_mode='sparse', shuffle=True
)
raw_val_gen = val_datagen.flow_from_directory(
    val_dir, target_size=(224, 224), batch_size=32, class_mode='sparse', shuffle=False
)

# 2. BULLETPROOF KERAS SEQUENCE MAPPING CLASS
# Maps: Grades [0, 1] -> 0 (Normal) | Grades [2, 3, 4] -> 1 (Osteoporosis)
class BroadBinaryMappingSequence(Sequence):
    def __init__(self, raw_generator):
        self.raw_gen = raw_generator
        self.target_size = (224, 224)

    def __len__(self):
        return len(self.raw_gen)

    def __getitem__(self, idx):
        batch_x, batch_y = self.raw_gen[idx]
        mapped_y = np.where(batch_y >= 2, 1.0, 0.0)
        return batch_x, mapped_y

    def on_epoch_end(self):
        if hasattr(self.raw_gen, 'on_epoch_end'):
            self.raw_gen.on_epoch_end()

    def reset(self):
        if hasattr(self.raw_gen, 'reset'):
            self.raw_gen.reset()

# Instantiating the clean mapped stream configurations
train_gen = BroadBinaryMappingSequence(raw_train_gen)
val_gen = BroadBinaryMappingSequence(raw_val_gen)

print("\nSuccessfully grouped 5 clinical grades into a broad 2-class spectrum sequence layout!")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

print("\nInitializing FULLY UNFROZEN EfficientNetB3 Core Engine...")

inputs = layers.Input(shape=(224, 224, 3))

# Integrated built-in scaling layer alignment
preprocessed_inputs = tf.keras.applications.efficientnet.preprocess_input(inputs)

# Load the core backbone engine
base_model = tf.keras.applications.EfficientNetB3(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3)
)

# CRITICAL UNFREEZE: Every single filter block is completely defrosted from the start
base_model.trainable = True

# High-density normalization head
x = base_model(preprocessed_inputs)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)

x = layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)

outputs = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(inputs, outputs)
print("Model initialized for full weight updates across all layers.")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Calculate optimization step timelines dynamically based on generator lengths
epochs = 30
steps_per_epoch = len(train_gen)
total_steps = steps_per_epoch * epochs
warmup_steps = steps_per_epoch * 3  # Gradual 3-epoch warmup to safely wake up weights

# Create a continuous mathematical schedule tensor
lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-4,     # Peak learning rate after warmup
    decay_steps=total_steps,
    alpha=1e-6,                     # Minimum baseline floor learning rate
    warmup_target=1e-4,
    warmup_steps=warmup_steps
)

# Compile using advanced AdamW optimizer tied directly to our custom decay curve
print("\nCompiling with full AdamW Cosine Decay Schedule...")
model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=6,
    restore_best_weights=True,
    verbose=1
)

print("\nCommencing Full Backbone Weight Adaptation Optimization Run...")
history = model.fit(
    train_gen,
    steps_per_epoch=steps_per_epoch,  # FIXED: Explicitly handle custom generator step definitions
    validation_data=val_gen,
    validation_steps=len(val_gen),
    epochs=epochs,
    callbacks=[early_stop]
)

# Simple fallback assignment so Cell 7 plotting curves continue to link cleanly
history_ft = type('Dummy', (object,), {'history': {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': [], 'acc': [], 'val_acc': []}})()

model.save("nutribone_binary_effnetB3_model.keras")
print("\nFull Optimization Run Complete and Model Saved!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# =====================================================================
# 1. PLOT METRIC TRENDS ACROSS BOTH STAGES
# =====================================================================
plt.figure(figsize=(14, 5))

# Combine Loss History
total_train_loss = list(history.history['loss']) + list(history_ft.history['loss'])
total_val_loss = list(history.history['val_loss']) + list(history_ft.history['val_loss'])

# Handle metric keys mapping safely
train_acc_key = 'accuracy' if 'accuracy' in history.history else 'acc'
val_acc_key = 'val_accuracy' if 'val_accuracy' in history.history else 'val_acc'

total_train_acc = list(history.history[train_acc_key]) + list(history_ft.history[train_acc_key])
total_val_acc = list(history.history[val_acc_key]) + list(history_ft.history[val_acc_key])

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(total_train_loss, label='Train Loss', color='royalblue')
plt.plot(total_val_loss, label='Val Loss', color='orange', linestyle='--')
plt.axvline(x=len(history.history['loss'])-1, color='gray', linestyle=':', label='Fine-Tuning Split')
plt.legend()
plt.title('Binary Multi-Stage Loss Progression')
plt.xlabel('Total Epochs')

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(total_train_acc, label='Train Acc', color='royalblue')
plt.plot(total_val_acc, label='Val Acc', color='orange', linestyle='--')
plt.axvline(x=len(history.history[train_acc_key])-1, color='gray', linestyle=':', label='Fine-Tuning Split')
plt.legend()
plt.title('Binary Multi-Stage Accuracy Progression')
plt.xlabel('Total Epochs')
plt.show()

# =====================================================================
# 2. BULLETPROOF 1-TO-1 ALIGNED PREDICTION BLOCK
# =====================================================================
print("\nExtracting binary production model from disk path...")
eval_model = tf.keras.models.load_model("nutribone_binary_effnetB3_model.keras")

# Reset our base raw validation generator for strict 1-by-1 sequential reads
raw_val_gen.shuffle = False
raw_val_gen.batch_size = 1
raw_val_gen.reset()

print("Calculating perfectly aligned validation metrics...")
y_true = []
preds = []

# FIXED: Extract both images and broad-mapped binary truth sequentially out of our generator
for i in range(len(raw_val_gen)):
    img, raw_lab = next(raw_val_gen)
    true_binary_label = 1 if raw_lab[0] >= 2 else 0
    y_true.append(true_binary_label)

    # Run prediction batch frame
    p = eval_model.predict(img, verbose=0)
    preds.append(p[0][0])

y_true = np.array(y_true)
preds = np.array(preds)

# CRITICAL BINARY FIX: Sigmoid maps to a continuous float array; check threshold at 0.5
y_pred = (preds > 0.5).astype(int).flatten()
class_names = ['Normal', 'Osteoporosis']

# Print results
print("\n--- Aligned Classification Report ---")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('True Aligned Binary Confusion Matrix')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping

# Configure Ensemble Depth
num_models = 3
ensemble_histories = []

# Ensure generators are set back to your optimal 224x224 shape
train_gen.target_size = (224, 224)
val_gen.target_size = (224, 224)

print(f"Commencing Deep Ensemble Training: Initializing {num_models} independent engines...")

for i in range(num_models):
    print(f"\n=======================================================")
    print(f" TRAINING ENSEMBLE MODEL VARIANT {i+1} / {num_models}")
    print(f"=======================================================")

    # 1. Reset global seeds dynamically per run to force diverse filter layouts
    tf.keras.utils.set_random_seed(42 + i)

    # 2. Build your maximum-performance 224x224 structure completely fresh
    inputs = layers.Input(shape=(224, 224, 3))
    preprocessed_inputs = tf.keras.applications.efficientnet.preprocess_input(inputs)

    base_model = tf.keras.applications.EfficientNetB3(
        include_top=False, weights='imagenet', input_shape=(224, 224, 3)
    )
    base_model.trainable = True # Fully unfrozen backbone

    x = base_model(preprocessed_inputs)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(1, activation='sigmoid')(x)
    variant_model = models.Model(inputs, outputs)

    # 3. Setup your winning Cosine Decay Schedule
    steps_per_epoch = len(train_gen)
    epochs = 25
    total_steps = steps_per_epoch * epochs
    warmup_steps = steps_per_epoch * 3

    lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
        initial_learning_rate=1e-4,
        decay_steps=total_steps,
        alpha=1e-6,
        warmup_target=1e-4,
        warmup_steps=warmup_steps
    )

    variant_model.compile(
        optimizer=tf.keras.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=1e-4),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    # Reset generators and shuffle train for unique presentation layouts
    train_gen.reset()

    # Set extended patience so every model cooks perfectly into its 96%+ training peak
    early_stop = EarlyStopping(
        monitor='val_loss', patience=12, restore_best_weights=True, verbose=1
    )

    # 4. Train Variant Model
    history = variant_model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=epochs,
        callbacks=[early_stop]
    )
    ensemble_histories.append(history)

    # Save variant strictly to disk path
    model_filename = f"nutribone_ensemble_variant_{i+1}.keras"
    variant_model.save(model_filename)
    print(f"Successfully locked and saved: {model_filename}")

print("\nAll Ensemble Components Trained Successfully!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("Loading saved broad-binary ensemble components...")
models_list = []
for i in range(num_models):
    models_list.append(tf.keras.models.load_model(f"nutribone_ensemble_variant_{i+1}.keras"))

# Reset raw val generator for strict sequential streaming
raw_val_gen.shuffle = False
raw_val_gen.batch_size = 1
raw_val_gen.reset()

print("Gathering cross-model broad predictions...")
y_true = []
all_preds = [[] for _ in range(num_models)]

# Pull images and build true binary mapping sequentially
for i in range(len(raw_val_gen)):
    img, raw_lab = next(raw_val_gen)
    true_bin = 1 if raw_lab[0] >= 2 else 0
    y_true.append(true_bin)

    # Process through each model variant
    for m_idx, eval_model in enumerate(models_list):
        pred = eval_model.predict(img, verbose=0)
        all_preds[m_idx].append(pred[0][0])

y_true = np.array(y_true)
all_preds = np.array(all_preds) # Shape: (3, total_samples)

# Ensemble voting consensus mean calculation
ensemble_probabilities = np.mean(all_preds, axis=0)
y_pred = (ensemble_probabilities > 0.5).astype(int)
class_names = ['Normal', 'Osteoporosis']

print("\n--- Broad Binary Ensemble Consensus Report ---")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Broad 2-Class Unified Confusion Matrix')
plt.show()
```
